```python
import pandas as pd
df = pd.read_csv('/content/drive/Shareddrives/BAX423/Project/Data generation/datageneration.csv')
df
```

```python
new_cols = ['Active Energy (kcal)', 'Apple Exercise Time (min)',
       'Apple Sleeping Wrist Temperature (degF)', 'Basal Energy Burned (kcal)',
       'Blood Oxygen Saturation (%)', 'Heart Rate [Min] (count/min)', \
       'Heart Rate [Max] (count/min)', 'Heart Rate [Avg] (count/min)', \
       'Heart Rate Variability (ms)', 'Respiratory Rate (count/min)',\
       'Resting Heart Rate (count/min)', 'Sleep Analysis [In Bed] (hr)',\
       'Sleep Analysis [Core] (hr)', 'Sleep Analysis [Deep] (hr)', \
       'Sleep Analysis [REM] (hr)', 'Sleep Analysis [Awake] (hr)', \
       'Step Count (count)', 'Walking Heart Rate Average (count/min)']
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df_new = df[new_cols]
```

```python
df_new=df_new.dropna(axis=0,how='any')
```

```python
df_new
```

```python
df_new.describe()
```

```python
from ctgan import CTGAN
```

```python
ctgan = CTGAN(epochs=10)
ctgan.fit(df_new)
```

```python
synthetic_df = ctgan.sample(200000)
synthetic_df.head()
```

```python
synthetic_df.describe()
```

```python
final_df = synthetic_df[(synthetic_df > 0).all(1)]
```

```python
final_df = final_df[final_df['Heart Rate [Max] (count/min)'] < 170] 
final_df = final_df[final_df['Sleep Analysis [In Bed] (hr)'] < 10 ]
final_df = final_df[final_df['Blood Oxygen Saturation (%)'] < 100 ]
final_df = final_df[final_df['Walking Heart Rate Average (count/min)'] < 130 ]
```
